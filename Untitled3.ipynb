```python
import pandas as pd 
import pandas as pd 
from matplotlib import pyplot as plt 
%matplotlib inline 
import seaborn as sns 
df = pd.read_csv('/content/zomato_modified.csv', index_col=0)

df=df.sample(frac=1)
#print(df)
df.head()
```

```python
df1=df.loc[df['Country'] == 'India']
df5=df1.loc[df['Aggregate rating'] >= 4]
df2=df1.loc[df['Has Online delivery'] == 'Yes']
df3=df1.loc[df['Has Online delivery'] == 'No']
df4=df.loc[df['Has Table booking'] == 'Yes']
df5.head()
```

```python
new = df["Cuisines"].str.split(",", n = 2, expand = True)
new['Rating']=df['Aggregate rating']
new['Cuisines']=new[0]
new['Country']=df['Country']
new['Average Cost for two'] = df['Average Cost for two']
df_new=df.loc[new['Cuisines'] == 'Italian']
df_new2=df_new.loc[df_new['Country']=='USA']
df_new2=df_new2.sample(frac=0.5)
df_new2['Average Cost for two']=74.36*df_new2['Average Cost for two']
f = plt.figure()
plt.rcParams.update({'font.size': 8})
f.set_figwidth(10)
f.set_figheight(3)
x=df_new2['Restaurant Name']
y=df_new2['Average Cost for two']
plt.bar(x,y)
plt.ylim(-1, 6000)
plt.xlim(0, 5)
mean2 = df_new2['Aggregate rating'].mean()
print(mean2)

#df_new['Average Cost for two']=74.36*df_new['Average Cost for two'] if df_new['City']=='USA'
#df_new['Average Cost for two']=df_new2['Average Cost for two']
#df_new2.head()
```
